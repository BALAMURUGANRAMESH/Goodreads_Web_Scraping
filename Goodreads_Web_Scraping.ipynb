2. Get the URL of the website you want to extract
* The HTTP GET method to retrieve the webpage's content, but it does so through Selenium's browser automation rather than a direct library call like requests.
Output: As you can see the HTML code, the data can be scraped from this site.

```python
from selenium import webdriver
driver = webdriver.Chrome()
url='https://www.goodreads.com/list/show/12362.All_Time_Favorite_Romance_Novels?page=1'
driver.get(url)
contents = driver.page_source
soup = bs(contents)
soup
```

3.3 Find the container in which the data we want to extract is stored.

```python
book_containers = soup.find_all('tr',itemtype ='http://schema.org/Book')
print(type(book_containers))
print(len(book_containers))
```

3.5 Extracting book title. Find the book title in the above HTML code.
We need to extract “Pride and Prejudice”

```python
name = first_book.find('a',class_="bookTitle")
name
```

We are getting the <a> class in which the book title is stored. But we only want name of the book and not the code.

So, we need to use text.strip() to extract the book title

```python
name = first_book.find('a',class_="bookTitle").text.strip()
name
```

3.6 Similarly, we will extract rest of the information for the first book.

```python
authors = first_book.find('a',class_="authorName").text.strip()
authors
```

Extracting scores for the first book.

```python
scoring = first_book.find('span',class_="greyText smallText uitext").text.strip().split()
scoring
```

As you can see that ratings and average ratings are stored in an array. But, we need them individually, so we need to extract them from the array.

```python
avg_scores=scoring[0]
rates = scoring[4]
print("averae sores:",avg_scores)
print("ratings", rates)
```

The last thing we need to extract are scores and votes.

```python
voted= first_book.find('span',class_="smallText uitext").text.strip().split()
scores=voted[1]
print("scores:",scores)
vote=voted[3]
print("votes:",vote)
```

5. We are going to follow the same steps, as we did for extracting first_book. But instead of directly printing, we are going to store the values in an array.

```python
page = 1
names = []
ratings = []
avgscores = []
author=[]
score=[]
votes=[]
while page != 51:
    url = f"https://www.goodreads.com/list/show/12362.All_Time_Favorite_Romance_Novels?page={page}"
    driver.get(url)
    content = driver.page_source
    soup = bs(content)
    book_containers = soup.find_all('tr', itemtype="http://schema.org/Book")
    for container in book_containers:
        if container.find('td', width= '100%') is not None:
            name = container.find('a',class_="bookTitle").text.strip()
            names.append(name)
            authors = container.find('a',class_="authorName").text.strip()
            author.append(authors)
            scoring = container.find('span',class_="greyText smallText uitext").text.strip().split()
            ascores=scoring[0]
            avgscores.append(ascores)
            rates = scoring[4]
            ratings.append(rates)
            voted= container.find('span',class_="smallText uitext").text.strip().split()
            scores=voted[1]
            score.append(scores)
            vote=voted[3]
            votes.append(vote)
    page = page + 1
```

As we have extracted the information, we can check whether the information is correct or not.

```python
names
```

6. We need to convert the above arrays into data frame, so that its convenient to use and read.

```python
df = pd.DataFrame({'book title': names,
'ratings': ratings,
'avg_score': avgscores,
'author': author,
'score': score,
 'votes':votes
})
df
```

7. We have successfully retrieved the data, you can also convert the data frame into csv file for further uses.

```python
df.to_csv("Goodreads_Data.csv")
```
